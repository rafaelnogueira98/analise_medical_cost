# 1. Carregando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# 2. Carregando dataset

```python
insurance = pd.read_csv("data/insurance.csv")
insurance.head()
```

```python
plt.scatter(data=insurance, x="bmi", y="charges")
plt.show()
```

```python
_ = plt.hist(data=insurance, x="age")
```

```python
# Como começamos:

smoke_counts = insurance["smoker"].value_counts()

# Tamanho da figura
plt.figure(figsize=(8,6))

# Criando o gráfico de barras com indice e contagem
plt.bar(
    smoke_counts.index, 
    smoke_counts.values, 
    color=["steelblue", "lightcoral"]
);

# Rotulo dos eixos x,y e o titulo
plt.xlabel("Smoker", fontsize=8)
plt.ylabel("Count", fontsize=8)
plt.title("Distribuição de Fumantes", fontsize=14)

# Remove as linhas do topo e da direita
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

plt.show()
```

```python
smoke_counts = insurance["smoker"].value_counts()

colors = ["steelblue", "lightcoral"] 

fig, ax = plt.subplots(figsize=(8,6))
ax.bar(smoke_counts.index, smoke_counts.values, color=colors)
ax.set_xlabel("Smoker", fontsize=8)
ax.set_ylabel("Count", fontsize=8)
ax.set_title("Distribuição de Fumantes", fontsize=14)

# Remove as linhas do topo e da direita
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.show()
```

```python
# Criando o gráfico de dispersão
fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(insurance["age"], insurance["charges"], c="red")

# Rotulo dos eixos x,y e o titulo
ax.set_xlabel("Age", fontsize=12)
ax.set_ylabel("Charges", fontsize=12)
ax.set_title("Gráfico de dispersão Idade vs Cobrança", fontsize=16)

# Remove as linhas do topo e da direita
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Alterar estilo dos gráficos do notebook
plt.style.use("ggplot")

# Exibe o gráfico
plt.show()
```

```python
# Cria o local e configuração da plot
fig, axes = plt.subplots(
    1, # numero de linhas
    2, # numero de colunas
    figsize=(12,6)
)

# Grafico de dispersao BMI x Charges na posição 0
# (linha 0, coluna 0)
axes[0].scatter(insurance["bmi"], insurance["charges"])
axes[0].set_xlabel("BMI")
axes[0].set_ylabel("Charges")
axes[0].set_title("BMI x Charges")

# Grafico de dispersao Age x Charges na posição 0
# (linha 0, coluna 1)

axes[1].scatter(insurance["age"], insurance["charges"])
axes[1].set_xlabel("Age")
axes[1].set_ylabel("Charges")
axes[1].set_title("Charges x Age")

# Ajustar espaço entre os plots
plt.tight_layout()

plt.show()
```

```python
# Calcular a média das cobranças por cada categoria de "sex"
mean_charges_sex = insurance.groupby("sex")["charges"].mean()

# Calcular a média das cobranças por cada categoria de "region"
mean_charges_region = insurance.groupby("region")["charges"].mean()

# Calcular a média das cobranças por cada categoria de "children"
mean_charges_children = insurance.groupby("children")["charges"].mean()

# Calcular a média das cobranças por cada categoria de "children"
mean_charges_smoker = insurance.groupby("smoker")["charges"].mean()

# Cores para as barras
colors = ["#2b8cbe", "#fe9929", "#31a354", "#a1d99b"]

# Criar uma figura com subplots
fig, axes = plt.subplots(2, 2, figsize=(12,8))

# Gráfico de barras: média das cobranças por sexo
mean_charges_sex.plot(kind="bar", ax=axes[0,0], color=colors)
axes[0,0].set_xlabel("Cobranças médias")
axes[0,0].set_ylabel("Sexo")
axes[0,0].set_title("Cobranças médias por sexo")

# Gráfico de barras: média das cobranças por região
mean_charges_region.plot(kind="bar", ax=axes[0,1], color=colors)
axes[0,1].set_xlabel("Cobranças médias")
axes[0,1].set_ylabel("Região")
axes[0,1].set_title("Cobranças médias por região")

# Gráfico de barras: média das cobranças por filhos
mean_charges_children.plot(kind="bar", ax=axes[1,0], color=colors)
axes[1,0].set_xlabel("Cobranças médias")
axes[1,0].set_ylabel("Filhos")
axes[1,0].set_title("Cobranças médias por filhos")

# Gráfico de barras: média das cobranças por fumantes
mean_charges_smoker.plot(kind="bar", ax=axes[1,1], color=colors)
axes[1,1].set_xlabel("Cobranças médias")
axes[1,1].set_ylabel("Fumantes")
axes[1,1].set_title("Cobranças médias por fumantes")

# Ajusta espaçamento entre os gráficos
plt.tight_layout()

plt.show()

```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))


graf_bar1 = axes[0][0].bar(insurance["sex"],insurance["charges"])
axes[0][0].spines["top"].set_visible(False)
axes[0][0].spines["right"].set_visible(False)

graf_bar2 = axes[0][1].bar(insurance["region"],insurance["charges"])
axes[0][1].spines["top"].set_visible(False)
axes[0][1].spines["right"].set_visible(False)

graf_bar3 = axes[1][0].bar(insurance["children"],insurance["charges"])
axes[1][0].spines["top"].set_visible(False)
axes[1][0].spines["right"].set_visible(False)

graf_bar4 = axes[1][1].bar(insurance["smoker"],insurance["charges"])
axes[1][1].spines["top"].set_visible(False)
axes[1][1].spines["right"].set_visible(False)

plt.tight_layout()

plt.show()
```

```python
# Criar um figura com duas linhas e duas colunas
fig, ax = plt.subplots(2, 2, figsize=(12,10))

# Primeira linha e primeira coluna: gráfico de smoker
sns.countplot(data=insurance, x="smoker", ax=ax[0,0])
ax[0,0].set_title("Smoker")

# Primeira linha e segunda coluna: gráfico de region
sns.countplot(data=insurance, x="region", ax=ax[0,1])
ax[0,1].set_title("Region")

# Segunda linha e primeira coluna: gráfico de children
sns.countplot(data=insurance, x="children", ax=ax[1,0])
ax[1,0].set_title("Children")

# Segunda linha e segunda coluna: gráfico de sex
sns.countplot(data=insurance, x="sex", ax=ax[1,1])
ax[1,1].set_title("Sex")

plt.tight_layout()

plt.show()
```

```python
# Cria um texto na posição 5 do eixo x e na posição 2 do eixo y, no tamanho 16
plt.text(
    5,
    2,
    "Texto Exemplo",
    size=16
)

# Definindo os limites dos eixos. O eixo x vai de 0 a 7, y vai de 0 a 4
plt.xlim(0,7)
plt.ylim(0,4)

# Exibe o gráfico
plt.show()
```

```python
# Contagem para cada tipo de região
region_types = insurance["region"].value_counts()

# Lista das regiões existentes
region_names = ["southwest", "southeast", "northwest", "northeast"]

# Cores personalizadas para as barras
colors = ["steelblue", "lightcoral", "mediumseagreen", "orange"]

# Cria figura com o eixo
fig, ax = plt.subplots(figsize=(8,6))

# Plota o gráfico de barras com as cores personalizadas
ax.bar(region_names, region_types, color=colors)

# Configurações dos rótulos dos eixos e o título
ax.set_xlabel("Região")
ax.set_ylabel("Contagem")
ax.set_title("Contagem de regiões")

# Adiciona contagem sobre cada barra
for i, v in enumerate(region_types.values):
    ax.text(i, v, str(v), ha="center", va="bottom")
    
plt.show()
```

```python
# Calcular a média de cobranças por sexo
mean_charges_sex = insurance.groupby("sex")["charges"].mean()

# Cores personalizadas por barra
colors = ["steelblue", "lightcoral"]

# Cria uma figura e um eixo
fig, ax = plt.subplots(figsize=(8,8))

# Plota o gráfico de barras com cores personalizadas
ax.bar(mean_charges_sex.index, mean_charges_sex.values, color=colors)

# Configurações dos rótulos dos eixos e o título
ax.set_xlabel("Sexo")
ax.set_ylabel("Média de cobranças")
ax.set_title("Média de cobranças por sexo")

# Adiciona média sobre cada barra
for i, v in enumerate(mean_charges_sex.values):
    ax.text(i, v, str(round(v,2)), ha="center", va="bottom", fontsize=12)
    
# Definir limites do eixo x e y
ax.set_xlim(-0.5,1.5)
ax.set_ylim(0, 16000)

# Calcula a diferença percentual entre os dois sexos
diff_perc = (mean_charges_sex["male"] - mean_charges_sex["female"])/mean_charges_sex["female"]*100

# Mensagem da diferença salarial
ax.text(
    1.2, 
    mean_charges_sex[1]+1000, 
    f"O salario dos homens é {diff_perc:.2f}% maior",
    bbox={"facecolor":"lightsteelblue",
          "alpha":1,
          "pad":0.7,
          "edgecolor":"none",
          "boxstyle":"round"},
    color="black",
    fontsize=11
)
    

    
plt.show()
```
